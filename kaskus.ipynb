```python
import requests
import pandas as pd
from bs4 import BeautifulSoup


import chromedriver_autoinstaller
from selenium import webdriver
```

```python
chromedriver_autoinstaller.install()
```

```python
browser = webdriver.Chrome()
```

```python
browser.get('https://www.kaskus.co.id/channel/9/food--travel')
```

```python
soup = BeautifulSoup(browser.page_source, 'lxml')
```

```python
soup
```

```python
soup.find('div', class_='column Fx(flexZero) W(400px) Mx(15px) Fl(start) jsStickyColumn')
```

```python
rows = soup.find_all('div', class_='P(15px) Bd(borderSolidLightGrey) Mb(15px) Bgc(c-white) Pos(r) Ov(h) jsThreadCard')
```

```python
data = []
for row in rows:
    community = row.find(class_='Ov(h) Tov(e) Whs(nw) Fz(14px)').text.strip()
    account = row.find(class_='C(c-secondary) Fw(500)').text.strip()
    pst = row.find('div', class_='Fw(500) Fz(18px) Mb(10px)')
    thread = pst.find(class_='C(c-primary)').text.strip()
    date = row.find(class_='C(c-tertiary)').text.strip()    
    
    kaskus = {
     'community': community,
        'account':account,
        'thread':thread,
        'date post':date,
    }
    data.append(kaskus)
```

```python
data
```

```python
pd.DataFrame(data).to_excel('kaskus.xlsx')
```
